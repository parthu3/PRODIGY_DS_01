```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("population.csv")
```

```python
data.head() #shows first 5 row details
```

```python
data.tail()# shows last 5 row details
```

```python
data.describe()
```

```python
#List of countries for the graph.
countries_plot = ['United States', 'Canada', 'Brazil', 'Japan', 'Germany', 'France']

#Filtering the data frame
filtered_data = data[data['Country Name'].isin(countries_plot)]

#Creating the bar chart
plt.bar (filtered_data['Country Name'], filtered_data['2020'])
plt.xlabel ('Country Name')
plt.ylabel ('Population in 2020')
plt.title('Population of selected countries in 2020')
plt.xticks(rotation=45, ha='right') #Rotating the x-axis labels for readability
plt.show()

```
